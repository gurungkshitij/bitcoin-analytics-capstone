# 1. Executive Summary

My analysis was focused on exploring the relationship between bitcoin coin metrics data and polymarket data. I found that Bitcoin's 2024-2025 run to $120k+ represents a fundamental shift in teh asset's lifecycle. While the previous runs were driven by Adoption (Active Addresses), the current era is defined by Institutional security (Hash) and Speculative Efficiency (Polymarket).
* Security vs Price: Hash Rate is now "decoupled" from price volatility, providing a more permanent security floor
* The Utility Gap: Active addresses have plateaued, suggesting Bitcoin is transitioning from a "Payment Network" to a "Settlement & Saving Layer"
* Crowd Wisdom vs. Noise: Polymarket activity has shifted from long-term thematic bets to high-frequency volatility plays, with market durations collapsing to near-zero.

# 2. Data Retrieval and Overview

Two main datasets were used:
* Coin Metrics BTC daily dataset for price and on-chain features
* Polymarket (markets, tokens, traders, odds, events, summary) for prediction market activities.

For Coin Metrics dataset, I converted the time to pandas datetime object and set it as index so that I could perform easy rolling calculation like 200 day Moving Average. We also calculated the MVRV quintile by applying 365-day rolling qcut to the raw ratio, essentially scoring the values form 1 to 5 against previous year. 

For Polymarket data,I used the Polars for its rapid handling of microsecond timestamps. I subseted the Bitcoin and Crypto centric theme buy filtering categories and questions usign keywords (BTC, Crypto, Business, Politics). I extracted daily sentiment and uncertainty by truncating high-frequency timestamps to a standard data grain to match our pricing data. I joined markets, Odds, Events, summary, with coin metrics dataset to create a master df enabling me to explore on-chain fundamental with real-time speculative psychology. 

# 3. Coin Metrics Exploration


I found that Bitcoin’s price is no longer purely speculative but is heavily anchored by Network Security and Capital Efficiency. The strongest long-term correlation exists between price and Hash Rate, which has moved in constant upward trajectory since 2010, persisting even during deep bear markets. In contrast, Active Addresses and Transaction Counts show a more cyclical relationship; they tend to lead price during bull runs but have recently plateaued, suggesting a shift from Bitcoin as a "medium of exchange" to a "settlement layer" for institutional capital.

The 200-Day Moving Average seem to provide a structural support during the 2024–2025 climb to $120k. Finally, the 1-Year ROI and Flow In USD metrics highlight Bitcoin’s extreme positive skew, where price is often stable for long periods before being driven by massive, high-conviction liquidity injections.
